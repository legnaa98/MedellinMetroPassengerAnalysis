# Medellin's Metro Passenger Data Analysis

Let us first import the data file using Pandas.

```python
# import modules
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib as mpl
sns.set()
pd.options.display.float_format = '{:,}'.format
```

# Data Cleaning
Now that we have the modules, we will import the data

```python
# import the data
df_pass = pd.read_csv("./resources/pasajeros_movilizados.csv", sep=";")
# take a look at the first 5 rows
df_pass.head()
```

## Rename column names for better understanding
We will rename the column names to have a more intuitive meaning.

e.g. **L-A_PAX_MOV** is just the number of passengers that used the Line A of the metro, so let's rename it to **LineA**

```python
df_pass.rename(columns={
    "L-A_PAX_MOV":"LineA", 
    "L-B_PAX_MOV":"LineB",
    "L-K_PAX_MOV":"LineK",
    "L-J_PAX_MOV":"LineJ",
    "L-H_PAX_MOV":"LineH",
    "L-M_PAX_MOV":"LineM",
    "L-L_PAX_MOV":"LineL",
    "L-TA_PAX_MOV":"LineTA",
    "L-1_PAX_MOV":"Line1",
    "L-2_PAX_MOV":"Line2",
    "C-3_PAX_MOV":"LineC3",
    "C6_PAX_MOV":"LineC6",
    "LÍNEA_O": "LineO"}, inplace=True)
df_pass.head()
```

## Change numeric values for correct data type
We examine the data types of each column and we find that the actual passenger values are floats, but the data was filled by spanish speaking people which means the dot (.) actually means a comma (or a thousand separator).

    e.g. 492.314 is not 492 point 314 but 492314, the point actually means a thousand separator.

We will build a function or a piece of code which handles this issue.

```python
df_pass.dtypes
```

```python
# columns to replace "." to ","
cols_to_replace = df_pass.columns.values[5:18]
# convert columns to string
df_pass[cols_to_replace] = df_pass[cols_to_replace].astype(str)
# apply the replace function to the strings and then convert into a float again
for col in cols_to_replace:
    df_pass[col] = df_pass[col].apply(lambda x: float(x.replace(".", "")))   
```

We can now see that our numbers are in the correct format

```python
df_pass.head()
```

## Examine missing of NaN values
We will now check for missing of nan values and plot how many entries are missing for each column.

```python
# find nan values and sum on each column then express as a percentage
nan_vals = pd.DataFrame(np.round(df_pass.isna().sum(axis=0)/df_pass.shape[0]*100, 3)).reset_index()
# rename columns of the nan values dataframe
nan_vals.columns = ["Column", "Nan Values (%)"]
# take only non zero nan values and sort in descending order
nan_vals = nan_vals[nan_vals["Nan Values (%)"] != 0].sort_values(by="Nan Values (%)", ascending=False)
# visualize results
plt.figure(figsize=(10,7))
chart = sns.barplot(data=nan_vals, x=np.arange(nan_vals.shape[0]), y="Nan Values (%)", hue="Column", dodge=False)
chart.axes.xaxis.set_visible(False)
```

As we can see **LineO** has more than 70% of its values as nan, this could mean that either there is missing data or that the line is not old enough so there isn't data for the earlier years.

We can validate this hypothesis by looking at the entries year by year, we see that there are entries since december 2019, and looking in the history of the Metro lines https://es.wikipedia.org/wiki/Metro_de_Medell%C3%ADn#L%C3%ADneas_tranviarias we can see that indeed this line started in december 2019.

# Exploratory Data Analysis (EDA)
## All lines passengers

```python
# get only lines columns
lines = df_pass.columns.values[5:18]

# create a copy to mess with the data
pax_by_month = df_pass.copy()
# summation of poassengers by month
pax_by_month = pd.DataFrame(pax_by_month.set_index(["AÑO", "NUM_MES"])[lines].sum(axis=1))
# rename default column 0 to Passengers
pax_by_month.columns = ["Passengers"]
# reset index in order to drop 2014 and 2015 data which have no monthly data
pax_by_month = pax_by_month.reset_index().dropna()
pax_by_month["YearMonth"] = pax_by_month.apply(lambda x: str(int(x["AÑO"])) + "-" + str(int(x["NUM_MES"])), axis=1)
pax_by_month.drop(columns=["AÑO", "NUM_MES"], inplace=True)
pax_by_month["YearMonth"] = pd.to_datetime(pax_by_month["YearMonth"], format="%Y-%m")

# plot the data
plt.figure(figsize=(15,7))
line_plot = sns.lineplot(data=pax_by_month, x="YearMonth", y="Passengers", marker="o")
plt.ticklabel_format(style='plain', axis='y')
line_plot.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
```

```python
pax_by_month.head(12)
```

## Passengers by line

```python
plt.figure(figsize=(15, 7))

# define color to iterate through
color = plt.cm.rainbow(np.linspace(0, 1, len(lines)))
# iterate through lines and colors
for line, c in zip(lines, color):
    # grab a copy from the original dataframe but only the current line and year, month columns
    df_tmp = df_pass[["AÑO", "NUM_MES", line]].copy()
    # set index to be year and month
    df_tmp = pd.DataFrame(df_tmp.set_index(["AÑO", "NUM_MES"]).sum(axis=1))
    # rename column of pax count to have the line name
    df_tmp.columns = ["Passengers " + line]
    # drop na values from year, month to plot only monthly data
    df_tmp = df_tmp.reset_index().dropna()
    # create a yearmonth column to correctly handle dates
    df_tmp["YearMonth"] = df_tmp.apply(
        lambda x: str(int(x["AÑO"])) + "-" + str(int(x["NUM_MES"])), axis=1
    )
    # we drop no longer needed columns
    df_tmp.drop(columns=["AÑO", "NUM_MES"], inplace=True)
    # convert yearmonth to date format
    df_tmp["YearMonth"] = pd.to_datetime(df_tmp["YearMonth"], format="%Y-%m")
    # plot data by line
    tmp_plot = plt.plot(
        df_tmp["YearMonth"],
        df_tmp["Passengers " + line],
        label=line,
        marker="o",
        markersize=1.5,
        color=c,
    )

plt.legend(loc="center left", bbox_to_anchor=(1, 0.5))
```

## Passengers by semester

```python
pax_by_semester = df_pass.groupby(["AÑO", "SEMESTRE"], as_index=False)[lines].sum()
pax_by_semester["Total Passengers"] = pax_by_semester[lines].sum(axis=1)
pax_by_semester["YearSemester"] = pax_by_semester.apply(lambda x: str(int(x["AÑO"])) + "-" + str(int(x["SEMESTRE"])), axis=1)

plt.figure(figsize=(15,7))
line_plot = sns.lineplot(data=pax_by_semester, x="YearSemester", y="Total Passengers", marker="o")
plt.ticklabel_format(style='plain', axis='y')
line_plot.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
```

```python
pax_by_year = df_pass.groupby(["AÑO"], as_index=False)[lines].sum()
pax_by_year["Total Passengers"] = pax_by_year[lines].sum(axis=1)
pax_by_year["Year"] = pax_by_year["AÑO"]

plt.figure(figsize=(15,7))
line_plot = sns.lineplot(data=pax_by_year, x="Year", y="Total Passengers", marker="o")
plt.ticklabel_format(style='plain', axis='y')
line_plot.yaxis.set_major_formatter(mpl.ticker.StrMethodFormatter('{x:,.0f}'))
```
